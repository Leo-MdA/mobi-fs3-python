**<a id="#">Task 3.1.2.</a>** Implement a mean filter:

```python
def meanfilter(img, size):
    result = numpy.zeros(img.shape)
    n = (size - 1) // 2
    for p in numpy.ndindex(img.shape):
        if p[0] >= n and p[0] < img.shape[0]-n and p[1] >= n and p[1] < img.shape[1] - n: 
            neighborhood_sum = 0
            for q in numpy.ndindex(size, size):
                neighborhood_sum = img[q[0] + p[0] - n, q[1] + p[1] - n]+ neighborhood_sum
            mean = neighborhood_sum / (size * size)
            result[p] = mean
    return result
```

## Non-linear filtering

**<a id="#">Task 3.2.1.</a>** Implement a median filter:

```python
def medianfilter(img, size):
    result = numpy.zeros(img.shape)
    n = (size) // 2
    for p in numpy.ndindex(img.shape):
        if p[0] >= n and p[0] < img.shape[0] -n and p[1] >= n and p[1] < img.shape[1] -n:
            list = []
            for q in numpy.ndindex(size, size):
                list.append(img[q[0] + p[0] - n, q[1] + p[1] - n])
            list.sort()
            median = list[(size) // 2]        
            result[p] = median            
    return result

plt.imshow(medianfilter(img, 3))
```
